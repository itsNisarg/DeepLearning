# Pytorch Basics

```python
import torch
```

```python
tensor = torch.arange(12, dtype=torch.float32)
```

```python
tensor.numel()      # number of elements in the tensor
```

```python
tensor.shape    # the tensor dimensions
```

```python
zero_t = torch.zeros(2,3,4)
```

```python
one_t = torch.ones(2,3,4)
```

```python
rand_t = torch.randn(2,3)
```

```python
exp_t = torch.exp(rand_t)
```

```python
torch.cat([zero_t, one_t], dim=0).shape, torch.cat([zero_t, one_t], dim=1).shape, torch.cat([zero_t, one_t], dim=2).shape
```

```python
x = torch.tensor([1,2,3]).reshape(-1,1)
y = torch.tensor([1,2]).reshape(1,-1)
```

```python
x, y
```

```python
id(x), id(y)
```

```python
x = x + y
```

```python
x, id(x)
```

```python
x += y
```

```python
x, id(x)
```

```python
A = torch.arange(12, dtype=torch.float32).reshape(3,4)
B = A.clone()
```

$A \odot B = A ⋆ B$ is called the Hadamard Product of Tensors

```python
print(A * B)
```

```python
A.sum(axis=0), A.sum(axis=1)    # reduction sum
```

```python
A.sum(axis=0, keepdims=True), A.sum(axis=1, keepdims=True)   # non-reduction sum
```

```python
A / A.sum(axis=1, keepdims=True)
```

```python
A.cumsum(axis=0), A.cumsum(axis=1)
```

```python
x = torch.tensor([1,2,3], dtype=torch.float32)
y = torch.ones(3)
```

```python
x, y, torch.dot(y, x)
```

```python
A.shape, B.shape
```

```python
y @ A, A.T @ y, A.T @ B, A @ B.T
```

```python
torch.norm(A)
```
